```python
#Using the columns Price, Area,No. of Bedrooms, Resale
df=df.drop('Location',axis=1)
df.drop(df.iloc[:,4:41],inplace=True, axis=1)
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x['Area'],y)
plt.title("Area and Beds v/s price in Hyderabad")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x.values,y.values)
```

```python
model.predict([[1250,6,0]])
```

```python
area=int(input("Enter the area : "))
beds=int(input("Enter the number of bedrooms : "))
resale=int(input("Is the house for resale : (enter 1 for resale and 0 for not a resale)"))
b=True
if resale!=0 and resale!=1:
    print("Invalid resale option")
    b=False
if beds<0:
    print("Invalid beds")
    b=False
if area<0:
    print("Invalid area")
    b=False
if b:
    print("The predicted price is ",model.predict([[area,beds,resale]]))
```
