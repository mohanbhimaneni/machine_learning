```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("D:/ml/datasets/Hyderabad.csv")
df
```

```python
#using only the price and area for linear regression
df.drop(df.iloc[:,2:41],inplace=True, axis=1)
df
```

```python
#splitting the data into two
x=df[['Area']]
y=df[['Price']]
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.title("Area v/s price in Hyderabad")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x.values,y.values)
```

```python
pred=model.predict(x.values)
```

```python
plt.scatter(x,y)
plt.plot(x,pred,color='red')
plt.title("Area v/s price in Hyderabad")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
i=int(input("Enter the area : "))
print("The predicted price is ",model.predict([[i]]))
```
